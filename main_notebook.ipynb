### Import some libaries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Declare a function

```python
def make_a_line(x):
    y = 10*x + 5
    return y
```

### Apply the function and plot it

```python
x = np.linspace(0,5,50)
y = make_a_line(x)
plt.scatter(x,y)
plt.show()
```
